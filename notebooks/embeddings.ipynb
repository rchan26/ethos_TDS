```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
seed = 2022

import pickle
```

```python
csv_file = "cleaned_EThOS_CSV_202210.csv"
df = pd.read_csv(csv_file, encoding="ISO-8859-2")
df.head()
```

# Spacy Embeddings

```python
import spacy
nlp = spacy.load('en_core_web_md')
```

```python
# obtain title embeddings
title_docs = list(nlp.pipe(df["Title"], n_process=6))
title_embeddings_spacy = [np.array(doc.vector) for doc in title_docs]
df["title_embeddings_spacy"] = title_embeddings_spacy
```

```python
# obtain title embeddings
abstract_docs = list(nlp.pipe(df["Abstract"], n_process=6))
abstract_embeddings_spacy = [np.array(doc.vector) for doc in abstract_docs]
df["abstract_embeddings_spacy"] = abstract_embeddings_spacy
```

# Huggingface Embeddings

```python
from sentence_transformers import SentenceTransformer
# model_name = "all-MiniLM-L6-v2" # alternative: "all-mpnet-base-v2", "all-distilroberta-v1", "all-MiniLM-L12-v2"
model_args = {"batch_size": 64,
              "show_progress_bar": True,
              "output_value": "sentence_embedding",
              "convert_to_numpy": True,
              "convert_to_tensor": False,
              "device": None,
              "normalize_embeddings": False}
```

From https://www.sbert.net/docs/pretrained_models.html, it looks like the `all-mpnet-base-v2` model generally provides the best sentence embeddings performance (based on 14 different tasks) and provides embeddings on a space of dimension 768.

```python
model_name = "all-mpnet-base-v2"
sentence_model = SentenceTransformer(model_name)

# # title embeddings
# title_embeddings_mpnet_base_v2 = sentence_model.encode(df["Title"], **model_args)
# df["title_embeddings_mpnet_base_v2"] = [title_embeddings_mpnet_base_v2[i] for i in range(len(df))]

# abstract embeddings
abstract_embeddings_mpnet_base_v2 = sentence_model.encode(df["Abstract"], **model_args)
df["abstract_embeddings_mpnet_base_v2"] = [abstract_embeddings_mpnet_base_v2[i] for i in range(len(df))]
```

```python
df.to_pickle("../EThOS_CSV_202210_with_embeddings.pkl")
```

However, the `all-MiniLM-L6-v2` model provides similar performance and is 5 times quicker. The embeddings are in 384 dimensions. But note that input text longer than 256 word pieces is truncated with this model as opposed to 384 above.

Would be interesting to see how much difference in performace we get between these three different ways to obtain embeddings...

```python
model_name = "all-MiniLM-L6-v2"
sentence_model = SentenceTransformer(model_name)

# title embeddings
title_embeddings_MiniLM_L6_v2 = sentence_model.encode(df["Title"], **model_args)
df["title_embeddings_MiniLM_L6_v2"] = [title_embeddings_MiniLM_L6_v2[i] for i in range(len(df))]

# abstract embeddings
abstract_embeddings_MiniLM_L6_v2 = sentence_model.encode(df["Abstract"], **model_args)
df["abstract_embeddings_MiniLM_L6_v2"] = [abstract_embeddings_MiniLM_L6_v2[i] for i in range(len(df))]
```

```python
df.to_pickle("../EThOS_CSV_202210_with_embeddings.pkl")
```
